```python
import matplotlib
import PIL
import scipy
from matplotlib import image as img
%matplotlib inline
from matplotlib import pyplot as plt

from scipy.cluster.vq import whiten
from scipy.cluster.vq import kmeans
import pandas as pd
```

```python
def get_single_colour_rgb(image_path, k):
    
    # Read
    image = img.imread(image_path)
    image.shape
    
    # Show
    plt.imshow(image)
    plt.show()
    
    # Format
    r = []
    g = []
    b = []
    
    for line in image:
        for pixel in line:
            temp_r, temp_g, temp_b, TEMP_X = pixel
            r.append(temp_r)
            g.append(temp_g)
            b.append(temp_b)
    
    df = pd.DataFrame({'red': r,'blue': b,'green': g})
    df['scaled_red'] = whiten(df['red'])
    df['scaled_blue'] = whiten(df['blue'])
    df['scaled_green'] = whiten(df['green'])

    # Get
    from scipy.cluster.vq import kmeans
    cluster_centers, distortion = kmeans(df[['scaled_red', 'scaled_green', 'scaled_blue']], 1)
    
    colors = []
    rgb = []
    r_std, g_std, b_std = df[['red', 'green', 'blue']].std()

    for cluster_center in cluster_centers:
        scaled_r, scaled_g, scaled_b = cluster_center
        colors.append((
        scaled_r * r_std,
        scaled_g * g_std,
        scaled_b * b_std
        ))
        rgb.append((
        scaled_r * r_std * 255,
        scaled_g * g_std * 255 ,
        scaled_b * b_std * 255
        ))

    plt.imshow([colors])
    plt.show()

    print(rgb[0])
    
    return rgb[0]
```

```python
import os
file_locations = [os.path.join(root, name)
                  for root, dirs, files 
                  in os.walk(r'img')
                  for name in files
                  if name.endswith((".png", ".jpg", ".PNG", ".JPG"))]
```

```python
k = 1
result = []

for image_path in file_locations:
    file_name = os.path.split(image_path)[-1]

    path = os.path.split(image_path)[-2]
    launch_name = os.path.split(path)[-1]

    path = os.path.split(path)[-2]
    year_name = os.path.split(path)[-1]

    file_name, launch_name, year_name

    trend_colour = file_name.split('-')[0]
    trend_index = (file_name.split('-')[1]).split('.')[0]
    base_colour = trend_colour.split('_')[-1]
    
    cluster_centers = get_single_colour_rgb(image_path,k)

    result.append((launch_name, year_name, base_colour, trend_colour, trend_index, image_path, cluster_centers))

result = pd.DataFrame(result)
result.columns = ['launch_name','year_name','base_colour','trend_colour','trend_index', 'image_path', 'rgb']
result.head()
```

```python
result.to_csv('results.csv')
```
